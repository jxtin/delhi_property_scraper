```python
# localhost:8000
import requests
```

```python
list_of_localities = requests.get("http://localhost:8000/localities").json()
```

```python
list_of_localities
```

```python
import concurrent.futures
```

```python
def scrape_locality(locality: str):
    return requests.get(f"http://localhost:8000/scrape?locality={locality}").json()
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
    # make a request to the server, wait for the response, and then print the response
    future_to_url = {executor.submit(scrape_locality, locality): locality for locality in list_of_localities}
    for future in concurrent.futures.as_completed(future_to_url):
        locality = future_to_url[future]
        try:
            data = future.result()
        except Exception as exc:
            print('%r generated an exception: %s' % (locality, exc))
        else:
            print('%r page is %d bytes' % (locality, len(data)))
```
